```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=10, n_features=1, n_informative=1, noise=80)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# Lets Apply OLS

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)

lr.coef_, lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X), color='red')
plt.show()
```

```python
# Applying Gradient Descent assuming m = 83.12857343

# And assuming the starting value for b = 0
b = 0
y_pred = (lr.coef_ * X + b)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X), color='green') # Sklearn OLS
plt.plot(X, y_pred, color='red') # When b = 0
plt.show()
```

```python
m = lr.coef_
b = 0
alpha = 0.001
n = y.shape[0]

for i in range(1000):
  for j in range(n):
     b = b - alpha * (-2 * (y[j] - (m*X[j] + b)))
```

```python
y_pred2 = (m*X+b)
```

```python
plt.scatter(X,y)
plt.plot(X, lr.predict(X), color='green') # Sklearn OLS
plt.plot(X, y_pred, color='red') # When b = 0
plt.plot(X, y_pred2, color='blue') # Calculated b using Gradient Descent
plt.show()
```
